```python
df = pd.read_csv('Dataset_Pemesanan_Hotel_1000Baris.csv', sep=';')
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print(df.columns.tolist())
```

```python
pemesanan_per_kota = df['kota'].value_counts()

pemesanan_per_kota.plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title('Jumlah Pemesanan per Kota')
plt.xlabel('Kota')
plt.ylabel('Jumlah Pemesanan')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
top_hotel = df['nama_hotel'].value_counts().head(10)

top_hotel.plot(kind='barh', figsize=(10,6), color='teal')
plt.title('Top 10 Hotel Paling Banyak Dipesan')
plt.xlabel('Jumlah Pemesanan')
plt.ylabel('Nama Hotel')
plt.gca().invert_yaxis()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Pastikan harga numerik
df['total_biaya'] = pd.to_numeric(df['total_biaya'], errors='coerce')

pendapatan_per_hotel = df.groupby('nama_hotel')['total_biaya'].sum().sort_values(ascending=False).head(10)

pendapatan_per_hotel.plot(kind='bar', figsize=(10,5), color='orange')
plt.title('Pendapatan Tertinggi per Hotel')
plt.xlabel('Nama Hotel')
plt.ylabel('Total Pendapatan (Rp)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['metode_pembayaran'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(7,7), shadow=True)
plt.title('Distribusi Metode Pembayaran')
plt.ylabel('')
plt.show()
```

```python
df['status_transaksi'].value_counts().plot(kind='bar', color='salmon', figsize=(6,4))
plt.title('Status Transaksi Pemesanan')
plt.xlabel('Status')
plt.ylabel('Jumlah')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Parsing waktu ke datetime
df['timestamp_pemesanan'] = pd.to_datetime(df['timestamp_pemesanan'], errors='coerce')

# Ambil bulan dan hitung
df['bulan'] = df['timestamp_pemesanan'].dt.to_period('M').astype(str)

pemesanan_bulanan = df['bulan'].value_counts().sort_index()

pemesanan_bulanan.plot(kind='line', marker='o', figsize=(10,5))
plt.title('Tren Jumlah Pemesanan per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Pemesanan')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Pastikan kolom numerik bersih
df['harga_per_malam'] = pd.to_numeric(df['harga_per_malam'], errors='coerce')

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='metode_pembayaran', y='harga_per_malam')
plt.title('Distribusi Harga per Malam berdasarkan Metode Pembayaran')
plt.xlabel('Metode Pembayaran')
plt.ylabel('Harga per Malam')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='status_transaksi', palette='Set2')
plt.title('Jumlah Transaksi per Status')
plt.xlabel('Status Transaksi')
plt.ylabel('Jumlah')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Ambil kolom numerik
num_df = df[['harga_per_malam', 'durasi_inap', 'jumlah_kamar', 'total_biaya']].copy()
num_df = num_df.apply(pd.to_numeric, errors='coerce')

# Buat korelasi
corr = num_df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title('Korelasi antar Variabel Numerik')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='harga_per_malam', bins=30, kde=True, color='purple')
plt.title('Distribusi Harga per Malam')
plt.xlabel('Harga per Malam (Rp)')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=df, x='metode_pembayaran', y='total_biaya', estimator='mean', ci='sd', palette='coolwarm')
plt.title('Rata-rata Total Biaya per Metode Pembayaran')
plt.xlabel('Metode Pembayaran')
plt.ylabel('Rata-rata Total Biaya')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.jointplot(data=df, x='harga_per_malam', y='total_biaya', kind='kde', fill=True, cmap='magma')
```

```python
df_paid = df[df['status_transaksi'] == 'Paid']

plt.figure(figsize=(10,6))
sns.boxplot(data=df_paid, x='metode_pembayaran', y='total_biaya')
plt.title('Distribusi Total Biaya (Paid Only)')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df[['harga_per_malam', 'jumlah_kamar', 'durasi_inap', 'total_biaya']], diag_kind='kde')
```

```python
top10 = df['nama_hotel'].value_counts().head(10).reset_index()
top10.columns = ['nama_hotel', 'jumlah']

fig = px.bar(top10, x='nama_hotel', y='jumlah', title='Top 10 Hotel Paling Banyak Dipesan',
             color='jumlah', color_continuous_scale='Tealgrn')
fig.update_layout(xaxis_title='Hotel', yaxis_title='Jumlah Pemesanan')
fig.show()
```

```python
df['harga_per_malam'] = pd.to_numeric(df['harga_per_malam'], errors='coerce')
df['total_biaya'] = pd.to_numeric(df['total_biaya'], errors='coerce')

fig = px.scatter(df, x='harga_per_malam', y='total_biaya', color='metode_pembayaran',
                 hover_data=['nama_hotel', 'status_transaksi'],
                 title='Hubungan Harga per Malam dan Total Biaya')
fig.update_layout(xaxis_title='Harga per Malam', yaxis_title='Total Biaya')
fig.show()
```

```python
df['timestamp_pemesanan'] = pd.to_datetime(df['timestamp_pemesanan'], errors='coerce')
df['bulan'] = df['timestamp_pemesanan'].dt.to_period('M').astype(str)
df_bulan = df['bulan'].value_counts().sort_index()

fig = px.line(df_bulan, x=df_bulan.index, y=df_bulan.values,
              title='Jumlah Pemesanan per Bulan', markers=True)
fig.update_layout(xaxis_title='Bulan', yaxis_title='Jumlah Pemesanan')
fig.show()
```
